### Analyzing the Data in the Dataset

#### Building the correlation matrix of the Dataset

```r
#bulding correlation matrix for the dataset
#!install.packages("Hmisc",repos='http://cran.us.r-project.org')
library(Hmisc)
#building correlation matrix
correlations <- rcorr(as.matrix(Standardized_ChurnData))
Variable_Correlations <- correlations$P
Variable_Correlations
```

```r
#fillling the diagonal values with "1" in the above correlation matrix
value <- 1
diag(Variable_Correlations) <- value
Variable_Correlations
```

```r
#converting the correlation matrix to dataframe to get highly correlated variables
correlations_DF <- as.data.frame(as.table(Variable_Correlations))
head(correlations_DF,10)
```

#### finding the top correlations among  variables in the dataset

```r
#finding the highly correlated variables in the dataframe with probability of 0.8
Top_correlations <- subset(correlations_DF,correlations_DF$Freq > 0.8)
head(Top_correlations,10)
```

```r
#removing the same attributes correlating to itself from the above dataframe
Top_correlations <- Top_correlations[!(Top_correlations$Freq == 1.0000000),]
Top_correlations_sorted <- as.data.frame(Top_correlations[order(Top_correlations$Freq),])
#function to delete every nth row in the dataframe
Nth.delete <- function(dataframe,n){ dataframe[-(seq(n,to = nrow(dataframe),by=n)),]}
#sorting the variables in the Ascending order according to their correlations greater than 0.8 probability
Top_correlations_sorted<- Nth.delete(Top_correlations,2)
Top_correlations_sorted <- Top_correlations_sorted[order(Top_correlations_sorted$Freq),]

head(Top_correlations_sorted,10)
```

#### Finding the correlations between independent variables and target variable

```r
#finding the correlation of the independent varaibles with the target varaible
corr_Target<- subset(correlations_DF,correlations_DF$Var2 == "Target") 
corr_Target_sorted <- corr_Target[which(corr_Target$Freq >0),]
corr_Target_sorted
```

### Selecting the Important variables in the Dataset(Dimensionality Reduction)

```r
#selecting the important varialbes in the dataset i.e., Dimensionality reduction
#using random forest algorithm filter for the selection of the important features 
#!install.packages("varSelRF",repos='http://cran.us.r-project.org' )
library(varSelRF)
ChurnData_subset <- ChurnData[,-c(18)]
Target_values <- ChurnData$target
Variables <- varSelRF(ChurnData_subset, Target_values, ntree = 200, ntreeIterat = 5,mtryFactor = 5,
              whole.range = T,c.sd = 0,returnFirstForest = T,vars.drop.frac =0.2)
```

```r
#storing variable importance in the dataframe in nascending order
Variables_matrix <- Variables$initialImportances
Variables_dataframe<- as.data.frame(Variables_matrix)
Variables_dataframe$variables <- rownames(Variables_dataframe)
rownames(Variables_dataframe)<-NULL
Variables_dataframe_sorted <- as.data.frame(Variables_dataframe[order(Variables_dataframe$MeanDecreaseAccuracy),])
head(Variables_dataframe_sorted)
```

```r
#adjusting the width and height of the figure in notebook 
library(repr)
options(repr.plot.width=7,repr.plot.height=4)
#plotting variable importance 
plot(Variables,which=1)
```

```r
#by the above figure we  can clearly see only six variables are have importance greater 0.1 rest all all have below 0.1
#selecting important six variables for model building
final_var <- c("s6.new.rev.p2.m2","s7.new.rev.p2.p6","s6.rtd.mou.p2.m2","s7.rtd.mou.p2.p6",
               "s3.og.mou.all.p2","s8.new.rev.p6","s7.rtd.mou.m1.m2")
#subsetting the data with selected six important variables
Standardized_ChurnData_sub <-Standardized_ChurnData[final_var]
Standardized_ChurnData_sub$Target <-ChurnData$target 
```

```r
#stratified sampling and spliting the data into train and test
library(caTools)
set.seed(456)
split = sample.split(Standardized_ChurnData_sub$Target, SplitRatio = 0.70)
train = subset(Standardized_ChurnData_sub, split==TRUE)
test = subset(Standardized_ChurnData_sub, split==FALSE)
#renaming target variable "0" as non churn and "1" as churn
levels(train$Target) <-c("not_churn","churn")
levels(test$Target) <-c("not_churn","churn")
```

### Building various ML Models to predict Churners and  Non Churn

####  Logistic Regression Model

```r
#building logistic regression model
Churn_LR<- glm(Target~.,data = train,family = binomial())
#predicting on the test data using LR Model
glmpredict.test <- predict(Churn_LR,test,type = "response")
```

```r
#creating ConfusionMatrix for LR Model
conf.test<- table(test$Target, glmpredict.test > 0.33)
accuracy <- (sum(diag(conf.test))/sum(conf.test))*100
cat("accuracy is --->", accuracy)
recall=(conf.test[2,2]/(conf.test[2,1]+conf.test[2,2]))*100  
cat("\nrecall is --->", recall) 
precision=(conf.test[2,2])/(conf.test[1,2]+conf.test[2,2])*100  
cat("\nprecision is --->", precision)
```

#### Buiding C5.0 Model(Decision Tree)

```r
library(C50)
C50_model =C5.0(Target~.,data=train,rules=T )
summary(C50_model)
```

```r
#creating ConfusionMatrix for C5.0 Model
conf.test=table(test$Target, predict(C50_model, newdata=test, type="class"))
accuracy <- (sum(diag(conf.test))/sum(conf.test))*100
cat("accuracy is --->", accuracy)
recall=(conf.test[2,2]/(conf.test[2,1]+conf.test[2,2]))*100  
cat("\nrecall is --->", recall) 
precision=(conf.test[2,2])/(conf.test[1,2]+conf.test[2,2])*100  
cat("\nprecision is --->", precision)
```

#### Buiding C.A.R.T Model (Decision Tree)

```r
library(rpart)
#building C.A.R.T Model
rpart_model<- rpart(Target~.,data=train,method = "class",
                    control = rpart.control(cp =0.001805054) )
cp_table<- rpart_model$cptable
```

```r
#creating ConfusionMatrix for C.A.R.T Model
conf.test<- table(test$Target,predict(rpart_model,newdata = test,type = "vector"))
accuracy <- (sum(diag(conf.test))/sum(conf.test))*100
cat("accuracy is --->", accuracy)
recall=(conf.test[2,2]/(conf.test[2,1]+conf.test[2,2]))*100  
cat("\nrecall is --->", recall) 
precision=(conf.test[2,2])/(conf.test[1,2]+conf.test[2,2])*100  
cat("\nprecision is --->", precision)
```

#### Building KNN(K Nearest Neighbours) Model

```r
library(class)
knn.train <- subset(train,select = -Target)
knn.test <- subset(test,select = -Target)
#building K.N.N Model
Churn_knn <- knn(knn.train,knn.test,train$Target,k=7)
```

```r
#creating ConfusionMatrix for K.N.N Model
conf.test <- table(Churn_knn,test$Target)
accuracy <- (sum(diag(conf.test))/sum(conf.test))*100
cat("accuracy is --->", accuracy)
recall=(conf.test[2,2]/(conf.test[2,1]+conf.test[2,2]))*100  
cat("\nrecall is --->", recall) 
precision=(conf.test[2,2])/(conf.test[1,2]+conf.test[2,2])*100  
cat("\nprecision is --->", precision)
```
